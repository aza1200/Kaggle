````python
import pandas as pd

test_df = pd.read_csv("./input/test.csv")
train_df = pd.read_csv("./input/train.csv")

# test_df.head()
# train_df.head()

# 기본적인 데이터 알기위해서 column 값 head 값 을 각각 출력해본다.
print(test_df.shape, train_df.shape)
print(test_df.columns.to_list())
print(train_df.columns.to_list())
````

````python
# 결측값(값이 존재하지 않는 csv파일들) 관찰 하기 위해 describe 사용
# 결측갯수 보고 

# ```
# Age   ( test 100개, train 200개)
# Cabin ( test 300개, train 700개 )
# Fare  ( test 1개, train 0개 )
# Embarked ( test 개, train 2개 )
# ```

test_df.describe()
# train_df.describe()

train_df.describe(include=['object'])
# test_df.describe(include=['object'])
train_df.info()
````

````python
# 성별에 따른 생존율 분석 

print(train_df[["Sex","Survived"]].groupby("Sex",as_index=False).mean())

# train_df[["Sex"]] -> dataframe 반환
# train_df["Sex"] -> Series 반환 (0~n 인덱스 가지는 리스트)

# female -> 0 // male-> 1 로 변환
train_df["Sex"] = train_df["Sex"].map({"female":1,"male":0})
test_df["Sex"] = test_df["Sex"].map({"female":1,"male":0})
train_df.head()
````

````python
# Pclass 에 따른 생존율 분석 
print(train_df[["Survived","Pclass"]].groupby("Pclass").mean())
# -> 걍 가만히 냅두겠음

# SibSp, Parch 에 따른 생존율 분석
# print(train_df[["Survived","SibSp"]].groupby("SibSp").mean())
# print(train_df[["Survived","Parch"]].groupby("Parch").mean())

# SibSp, Parch 의 변수에 따른 생존율 분석 
# print(train_df[["Survived","Parch","SibSp"]].groupby(["Parch","SibSp"]).mean())

# 혼자 온사람이 대체적으로 같이 온사람보다 생존율이 높으므로 IS_Alone 이라는 변수 새롭게 생성 

train_df["Is_Alone"] = (train_df["Parch"] + train_df["SibSp"]).apply(lambda x: int(x>0))
test_df["Is_Alone"] = (test_df["Parch"] + test_df["SibSp"]).apply(lambda x: int(x>0))
# print(train_df["Is_Alone"])
# train_df.head()

test_df.head()
train_df.head()
````

````python
train_df[["Survived","Embarked"]].groupby("Embarked").mean()

# test_df["Sex"] = test_df["Sex"].map({"female":1,"male":0})
# C = 0 Q = 1 S =2 로 치환 하자 
train_df["Embarked"] = train_df["Embarked"].map({"C":0,"Q":1,"S":2})
test_df["Embarked"] = test_df["Embarked"].map({"C":0,"Q":1,"S":2})

````

````python
train_df["Name_Cleaned"] = train_df["Name"].str.extract(r'([A-Za-z]+)\.', expand=False)
test_df["Name_Cleaned"] = test_df["Name"].str.extract(r'([A-Za-z]+)\.', expand=False)
````
