````python
import pandas as pd

test_df = pd.read_csv("./input/test.csv")
train_df = pd.read_csv("./input/train.csv")
````

````python
train_df["Sex"] = train_df["Sex"].map({"female":1,"male":0})
test_df["Sex"] = test_df["Sex"].map({"female":1,"male":0})

train_df["Is_Alone"] = (train_df["Parch"] + train_df["SibSp"]).apply(lambda x: int(x>0))
test_df["Is_Alone"] = (test_df["Parch"] + test_df["SibSp"]).apply(lambda x: int(x>0))

train_df["Embarked"] = train_df["Embarked"].map({"C":0,"Q":1,"S":2})
test_df["Embarked"] = test_df["Embarked"].map({"C":0,"Q":1,"S":2})

train_df["Name_Cleaned"] = train_df["Name"].str.extract(r'([A-Za-z]+)\.', expand=False)
test_df["Name_Cleaned"] = test_df["Name"].str.extract(r'([A-Za-z]+)\.', expand=False)
````

````python

# ```
# Age   ( test 100개, train 200개)
# Cabin ( test 300개, train 700개 )
# Fare  ( test 1개, train 0개 )
# Embarked ( test 개, train 2개 )
# ```

# for tmp in train_df["Name_Cleaned"]:
#     print(tmp)

tmp_df = train_df[["Survived","Name_Cleaned"]].groupby("Name_Cleaned").mean().reset_index()
tmp_dict = dict(zip(tmp_df.Name_Cleaned, tmp_df.Survived))


train_df["Name_Renual"] = (train_df["Name_Cleaned"]).apply(lambda x: int(tmp_dict[x]>=0.5) if x in tmp_dict else 0)
test_df["Name_Renual"] = (test_df["Name_Cleaned"]).apply(lambda x: int(tmp_dict[x]>=0.5) if x in tmp_dict else 0)

test_df.head()
````

````python
def age_band(x):
    if x<=16:
        return 1
    elif x<=32:
        return 2
    elif x<=48:
        return 3
    elif x<=64:
        return 4
    else:
        return 5

train_df["Age"].fillna(train_df["Age"].median(),inplace=True)
test_df["Age"].fillna(test_df["Age"].median(),inplace=True)

train_df["Age"].fillna(train_df["Age"].median(),inplace=True)
test_df["Age"].fillna(test_df["Age"].median(),inplace=True)

train_df["Embarked"].fillna(train_df["Embarked"].median(),inplace=True)
test_df["Embarked"].fillna(test_df["Embarked"].median(),inplace=True)

train_df["Fare"].fillna(train_df["Fare"].median(),inplace=True)
test_df["Fare"].fillna(test_df["Fare"].median(),inplace=True)

train_df["Cabin"].fillna(train_df["Cabin"].mode()[0],inplace=True)
test_df["Cabin"].fillna(test_df["Cabin"].mode()[0],inplace=True)


train_df["Age_Renual"] = (train_df["Age"]).apply(age_band)
test_df["Age_Renual"] = (test_df["Age"]).apply(age_band)

train_df.head()
````

````python
train_df = train_df.drop(["Name","Age","PassengerId","SibSp","Parch","Ticket","Fare","Cabin","Name_Cleaned"],axis=1)
train_df.head()
````

````python
test_df = test_df.drop(["Name","Age","PassengerId","SibSp","Parch","Ticket","Fare","Cabin","Name_Cleaned"],axis=1)

test_df.head()
````

````python
X_train = train_df.drop(["Survived"], axis=1)
Y_train = train_df["Survived"]

X_test  = test_df

X_train.shape, Y_train.shape , X_test.shape
````

````python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log

Y_pred
````
